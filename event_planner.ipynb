```python
SHEDULE_TABLE_PATH = 'PATH\\shedule.xlsx'
DIST_TABLE_PATH = 'PATH\\dist.xlsx'

# Готвим таблицу непосредственно с расписанием сотрудников
def prepare_data(path_to_table):
    table = pd.read_excel(path_to_table)
    table = table.transpose()
    table = table.rename(columns=table.iloc[0])
    table = table.iloc[1:].reset_index()
    table['index'] = pd.to_datetime(table['index'], format='%d.%m.%Y')
    
    display(table)
    return table
    
shedule_table = prepare_data(SHEDULE_TABLE_PATH)
dist_table = prepare_data(DIST_TABLE_PATH)
```

```python
# Создаем список из дат и интервалов сотрудника
def create_logins_shedules(shedule_table):
    shedule_tables = list()
    for column in shedule_table.columns:
        if column != 'index':
            shedule_login = shedule_table[['index', column]]
            shedule_login = shedule_login.dropna(axis='index', how='any', inplace=False)
            shedule_tables.append(shedule_login)
        else:
            continue
    return shedule_tables
    
# Вычисляем отдельно взятые часы и даты из строкового написания интервала
def get_hours(interval_str, date, login):
    try:
        working_dates_hours = dict()
        start, end = interval_str.split(' - ')
        start_hour = int(start.split(':')[0])
        end_hour = int(end.split(':')[0])

        if start_hour > end_hour:
            next_date = date + pd.Timedelta(days=1)
            next_date_start_hour = 0
            next_date_end_hour = end_hour
            end_hour = 24 

            next_day_hours = [f"{hour:02d}:00" for hour in range(next_date_start_hour, next_date_end_hour)]
            working_dates_hours.update({next_date: next_day_hours})

            this_day_hours = [f"{hour:02d}:00" for hour in range(start_hour, end_hour)]
            working_dates_hours.update({date: this_day_hours})
        else: 
            this_day_hours = [f"{hour:02d}:00" for hour in range(start_hour, end_hour)]
            working_dates_hours.update({date: this_day_hours})
    except Exception as e:
        print(f"Ошибка на логине {login}, {e}")    
        
    
    return working_dates_hours
    
# Находим максимальные профициты в даты и интервалы, когда работает сотрудник
def define_proficit_hours(working_dates_hours, login, dist_table):
    max_proficit_hours = dict()
    common_dates = list()
    
    # Сохраним все уникальные даты из рабочего расписания
    working_dates = set(working_dates_hours.keys())
    
    # Проверим все даты из dist_table
    for date in dist_table.index:
        if date in working_dates:
            common_dates.append(date)
    
    # Обрабатываем только даты, которые есть в обоих наборах
    for date in common_dates:
        try:
            date_dt = pd.to_datetime(date)
            
            bf_d = working_dates_hours[date]
            filtered_df = dist_table.loc[dist_table.index.isin([date_dt]), dist_table.columns.isin(bf_d)]
            
            if not filtered_df.empty:
                max_value = max(filtered_df.max())
                proficit_hour = filtered_df.columns[filtered_df.isin([max_value]).any()].tolist()[0]
                max_proficit_hours.update({date_dt: [proficit_hour, max_value]})
            else:
                print(f"Нет данных для даты {date}")
        except Exception as e:
            print(f"Ошибка при обработке даты {date}: {e}")
    
    return max_proficit_hours

# Находим наибольший профицит из полученного в define_proficit_hours словаря профицитов
def define_max_proficit(max_proficit_hours):
    max_value = max(max_proficit_hours.values(), key=lambda x: x[1])
    max_date = max(max_proficit_hours.keys(), key=lambda k: max_proficit_hours[k][1])

    hour = max_value[0]
    max_value = max_value[1]
    
    print(f"Дата: {max_date}")
    print(f"Час: {hour}")
    print(f"Максимальное значение: {max_value}")

    return max_date, hour

shedule_tables = create_logins_shedules(shedule_table)
dist_table = dist_table.reset_index().set_index('index')

final_login_dist_table = dict()

for login_shedule in shedule_tables:
    login = login_shedule.columns[1]
    work_intervals = dict(zip(login_shedule['index'], login_shedule[login_shedule.columns[1]]))
    working_dates_hours = dict()
    
    # Накапливаем все интервалы для одного сотрудника
    for date, interval in work_intervals.items():
        hours = get_hours(interval, date, login)
        working_dates_hours.update(hours)

    try:
        max_proficit_hours = define_proficit_hours(working_dates_hours, login, dist_table)
        
        max_date, hour = define_max_proficit(max_proficit_hours)
        
        print(login)
        final_login_dist_table.update({login: [max_date, hour]})
        
        dist_table.loc[max_date, hour] = dist_table.loc[max_date, hour]-1
    except Exception  as err:
        print(err)
        continue
```

```python
# Экспорт в Эксель
final_login_dist_table = pd.DataFrame.from_dict(final_login_dist_table, orient='index')
final_login_dist_table.columns = ['дата', 'время']

dist_table = dist_table.drop(columns='level_0', axis=1, inplace=False)
dist_table.to_excel('Dist_final.xlsx', index_label='date')

final_login_dist_table.to_excel('Распределение по профициту.xlsx', index_label='login')
```
